```python
selected_features = []
for idx in range(1, len(train_df.columns)):
    try:
        feature = train_df.columns[idx]
        corr = train_df[feature].corr(y)
        print(f'Feature {feature} correlation with y: {corr}')
        selected_features.append(feature)
    except TypeError as error:
        print(f'Feature {feature}')

#print(f'Selected features: {selected_features}')
for feature in selected_features:
    #print(f'mean for {feature}: {train_df[feature].mean()}')
    train_df[feature] = train_df[feature].fillna(train_df[feature].mean())
    
train_df = train_df.fillna(0) # other fields
```

```python
selected_features.append('MSZoning')
selected_features.append('Street')
selected_features.append('Alley')
selected_features.append('LotShape')
X = train_df[selected_features]
X = pd.get_dummies(X)
X
```

```python
train_df['LotShape'].unique()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.33, random_state=42)

```

```python
model = LinearRegression().fit(X_train, y_train)
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
